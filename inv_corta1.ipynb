```python
# The following Pandas investigation reunites several pandas functions used to treat a dataset of values
# an retrieve useful information from it.


# Libraries used in this example
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset import
dataset = pd.read_csv("precipitation.csv", delimiter = ';', names = ['date', 'month', 'year', 'value','fl_miss', 'fl_cmiss', 'country', 'indicator'])
```

```python
# How to take 5 random samples of the dataset
samples = dataset.sample(5)
samples
```

```python
# How to select 2 columns
date_and_value = dataset[['date', 'value']]
```

```python
# How to group all dataset by a 
monthly_mean = dataset.groupby('month').value.mean()
monthly_mean
```

```python
# How to calculate the min value of a column 
dataset.value.min()
```

```python
# How to calculate the max value of a column 
dataset.value.max()
```

```python
# How to calculate the mean value of a column 
dataset.value.median()
```

```python
# How to make a bar graph the dataset grouped by  a column of interest
bar_graph = dataset.groupby('month').value.mean().plot(kind='bar')
bar_graph.set_title("Montly mean precipitation in Costa Rica")
# bar_graph.set_xlabel("Month")
# bar_graph.set_ylabel("Value")
```

```python
# How to make a scatter graph the dataset grouped by  a column of interest
scatter_graph = dataset.plot.scatter(x='year', y='value')
scatter_graph.set_title("Precipitation in Costa Rica")
scatter_graph.set_xlabel("Year")
scatter_graph.set_ylabel("Value")
```

```python
# How to make a scatter graph the dataset grouped by  a column of interest
pie_graph = dataset.groupby('year').value.mean().plot.pie(figsize=(5, 5))
pie_graph.set_title("Precipitation in Costa Rica from 2010 to 2015")
pie_graph.set_xlabel("Year")
```

```python
# How to drop from the dataset any missing data
dataset_wo_nan = dataset.dropna()
dataset_wo_nan
```

```python
# How to drop from the dataset rows that have at least 1 missing data
dataset.dropna(thresh=1)
```

```python
# How to refill the dataset missing data with a mean value
dataset.fillna(dataset.mean())
```
